### 3.2 CAPM

The Capital Asset Pricing Model (CAPM) asserts that an asset (or portfolio’s) expected excess return is completely a function of its beta to the equity market index (`SPY`, or in this case, `MKT`).

Specifically, it asserts that, for any excess return $\tilde{r}^i$, its mean is proportional to the mean excess return of the market, $\tilde{r}^{mkt}$, where the proportionality is the regression beta of $\tilde{r}^i$ on $\tilde{r}^{mkt}$:

$$
\mathbb{E}[\tilde{r}^i_t] = \beta^{i,mkt} \, \mathbb{E}[\tilde{r}^{mkt}_t]
$$

Let’s examine whether that seems plausible.

For each of the $n = 25$ test portfolios, run the CAPM time-series regression:

$$
\tilde{r}^i_t = \alpha^i + \beta^{i,mkt} \tilde{r}^{mkt}_t + \varepsilon^i_t
$$

So you are running 25 separate regressions, each using the $T$-sized sample of time-series data.

#### Tasks

- Report the betas and alphas for each test asset.  
- Report the mean absolute error (MAE) of the CAPM:  

$$
  MAE = \frac{1}{n} \sum_{i=1}^{n} |\alpha_i|
$$

If the CAPM were true, what would we expect of the MAE?

__It should be near zero__

### Additional Questions

- Report the estimated $\beta^{i,mkt}$, Treynor Ratio, $\alpha^i$, and Information Ratio for each of the $n$ regressions.
- If the CAPM model were true, what would be true of the Treynor Ratios, alphas, and Information Ratios?

1. __Alphas__ to be approximately zero.
2. __Treynor ratios__ to be the same across all regressions, with values similar to the expected excess return of the market factor.
3. __Information ratio__ to be close to zero.

```python
independent_variables = "Mkt-RF"

def full_capm_analysis(df_returns, market_factor, asset_list, freq=12, include_const=True):
    """
    Perform a complete two-step CAPM analysis directly inside a notebook.

    Stage 1: Time-series regressions for each asset → estimate Alpha and Beta.
    Stage 2: Cross-sectional regression → test CAPM prediction between average returns and betas.

    Parameters
    ----------
    df_returns : pd.DataFrame
        DataFrame containing excess returns for all assets and factors.
        Must include the market factor column specified in `market_factor`.

    market_factor : list
        List with the name(s) of the market factor(s) (e.g., ["Mkt-RF"]).

    asset_list : list
        List of asset names (column names in df_returns) to test.

    freq : int, optional (default=12)
        Frequency used to annualize values (e.g., 12 for monthly data).

    include_const : bool, optional (default=True)
        Whether to include an intercept (alpha) in regressions.

    Returns
    -------
    df_stage1 : pd.DataFrame
        Time-series regression results per asset (Alpha, Beta, R², std_e).

    MAE_alpha : float
        Mean absolute value of alphas across all assets.

    df_cross : pd.DataFrame
        Cross-sectional regression results (Nu, Lambda).
    """

    # -------------------------------------------------------
    # 1️⃣ Stage 1: Time-series regressions (per asset)
    # -------------------------------------------------------
    stage1_results = []

    for asset in asset_list:
        X = df_returns[market_factor]
        if include_const:
            X = sm.add_constant(X)
        y = df_returns[[asset]]

        model = sm.OLS(y, X).fit()

        # Alpha and Beta
        alpha = model.params["const"] * 100 * freq if include_const else 0
        beta = model.params[market_factor[0]] * 100
        r2 = model.rsquared * 100
        std_e = np.std(model.resid, ddof=1) * 100 * mth.sqrt(freq)

        df_temp = pd.DataFrame({
            "Alpha": [alpha],
            f"Beta_{market_factor[0]}": [beta],
            "R2": [r2],
            "std_e": [std_e]
        }, index=[asset])

        stage1_results.append(df_temp)

    df_stage1 = pd.concat(stage1_results)
    MAE_alpha = df_stage1["Alpha"].abs().mean()

    # -------------------------------------------------------
    # 2️⃣ Stage 2: Cross-sectional regression
    # -------------------------------------------------------
    # Prepare dataset: mean returns vs betas
    df_betas = df_stage1.filter(like="Beta_")
    df_betas.columns = [c.replace("Beta_", "") for c in df_betas.columns]
    df_cross = pd.concat([
        pd.DataFrame(df_returns[asset_list].mean(), columns=["mean"]),
        df_betas / 100
    ], axis=1).dropna()

    X_cross = df_cross[[market_factor[0]]]
    if include_const:
        X_cross = sm.add_constant(X_cross)
    y_cross = df_cross[["mean"]]

    cross_model = sm.OLS(y_cross, X_cross).fit()

    # Extract results: Nu (intercept) and Lambda (slope)
    Nu = cross_model.params["const"] * 100 * 12 if include_const else 0
    Lambda = cross_model.params[market_factor[0]] * 100
    R2_cross = cross_model.rsquared * 100

    df_cross = pd.DataFrame({
        "Nu": [Nu],
        f"Lambda_{market_factor[0]}": [Lambda],
        "R2_cross": [R2_cross]
    })

    # -------------------------------------------------------
    # 3️⃣ Return results
    # -------------------------------------------------------
    return df_stage1, MAE_alpha, df_cross
```

```python
df_excess_returns = pd.concat([df_excess_returns_total, df[["Mkt-RF"]]], axis = 1)

df_stage1, MAE_alpha, df_cross = full_capm_analysis(
    df_excess_returns,     # DataFrame con factores y activos
    ["Mkt-RF"],                  # Columna del factor de mercado
    df_excess_returns_total.columns,   # Lista de activos (portafolios)
    freq=12                      # Para anualizar
)

print("MAE_alpha:", MAE_alpha)
display(df_stage1.head())
```

### 3.3. Cross-sectional Estimation

Let’s test the CAPM directly. We already have what we need:

- The dependent variable, (y): mean excess returns from each of the $n = 25$ portfolios.  
- The regressor, (x): the market beta from each of the $n = 25$ time-series regressions.

Then we can estimate the following equation:

$$
\underbrace{\mathbb{E}[\tilde{r}^i]}_{n \times 1 \text{ data}} 
= 
\underbrace{\eta}_{\text{regression intercept}} 
+ 
\underbrace{\beta^{i, mkt}}_{n \times 1 \text{ data}} 
\underbrace{\lambda_{mkt}}_{\text{regression estimate}} 
+ 
\underbrace{\upsilon}_{n \times 1 \text{ residuals}}
$$

### Notes

- We use sample means as estimates of $\mathbb{E}[\tilde{r}^i]$.  
- This is a *weird* regression! The regressors are the betas from the time-series regressions we already ran.  
- This is a single regression, where we are combining evidence across all $n = 25$ series. Thus, it is a *cross-sectional regression*.  
- The notation emphasizes that the intercept is different from the time-series $\alpha$ and that the regressor coefficient is different from the time-series betas.

### Report

- The R-squared of this regression.  
- The intercept, $\eta$.  
- The regression coefficient, $\lambda_{mkt}$.
